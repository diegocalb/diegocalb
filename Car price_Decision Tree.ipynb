# **Car prices prediction with a decision tree**

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import classification_report, confusion_matrix
```

# **EDA**

```python
carPrice = pd.read_csv("CarPrice_A.csv")
carPrice.head()
```

```python
carPrice.info()
```

```python
carPrice.describe()
```

## **Select the features for the model ("enginesize", "stroke", "horsepower", "peakrpm", "citympg", "highwaympg")**

```python
y=carPrice.price
features=["enginesize", "stroke", "horsepower", "peakrpm", "citympg", "highwaympg"]
X=carPrice[features]
```

# **Train and test data**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
carPriceModel = DecisionTreeRegressor()
carPriceModel.fit(X_train, y_train)
y_pred = carPriceModel.predict(X_test)

# get predicted prices on validation data
print(mean_absolute_error(y_test, y_pred))
```

```python
df1=pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
df1
```
