### Hacer una bÃºsqueda sobre un usuario CAHuracan

```python
number_tweets=50
tweets=[]
likes=[]
time=[]

cursor= tweepy.Cursor(api.user_timeline, id="CAHuracan", tweet_mode="extended").items(number_tweets)
for i in cursor:
    tweets.append(i.full_text)
    likes.append(i.favorite_count)
    time.append(i.created_at)
```

### Armar el DataFrame

```python
df=pd.DataFrame({"Tweet":tweets, "Likes":likes, "Time":time})
df
```

### Eliminar tweets que no son de interÃ©s

```python
df1=df[~df.Tweet.str.contains("PT")]
df2=df1[~df1.Tweet.str.contains("ST")]
df2=df2.reset_index(drop=True)
```
